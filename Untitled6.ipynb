```python
def find_shp(stem):
    hits = glob.glob(f"**/{stem}.shp", recursive=True)
    if not hits:
        raise FileNotFoundError(
            f"{stem}.shp not found under {os.getcwd()} — drag the Shapefiles "
            "folder into your notebook folder in the JupyterLab file browser.")
    return hits[0]

mori = gpd.read_file(find_shp("Morigaon")).to_crs(4326)
siva = gpd.read_file(find_shp("Sivsagar")).to_crs(4326)

targets = gpd.GeoDataFrame(
    {"shapeName": ["Morigaon", "Sivasagar"]},
    geometry=[mori.geometry.iloc[0], siva.geometry.iloc[0]],
    crs=4326,
)
OUT = "footprints"; os.makedirs(OUT, exist_ok=True)
for n, g in zip(targets["shapeName"], targets.geometry):
    print(n, "bounds:", [round(v, 3) for v in g.bounds])
```

```python
import geopandas as gpd, subprocess, sys, shutil, os

BND_FILE = "Assam_only.shp"
NAME_COL = "shapeName"

OUT = "footprints"; os.makedirs(OUT, exist_ok=True)
PQ  = "overture_parquet"; os.makedirs(PQ, exist_ok=True)
OVERTURE = [shutil.which("overturemaps")] if shutil.which("overturemaps") \
           else [sys.executable, "-m", "overturemaps"]

adm = gpd.read_file(BND_FILE)
adm = adm.set_crs(4326) if adm.crs is None else adm.to_crs(4326)
print("districts:", len(adm))

def safe(s): return "".join(c if c.isalnum() else "_" for c in str(s)).strip("_")

done, skipped, failed = [], [], []
for i, row in adm.reset_index(drop=True).iterrows():
    name = str(row[NAME_COL]); tag = safe(name)
    if os.path.exists(f"{OUT}/{tag}_osm.shp"):
        skipped.append(name); continue
    try:
        poly = row.geometry.buffer(0)
        w, s, e, n = poly.bounds
        pq = f"{PQ}/{tag}.parquet"
        if not os.path.exists(pq):
            print(f"[{i+1}/{len(adm)}] downloading {name} ...")
            subprocess.run(OVERTURE + ["download", f"--bbox={w},{s},{e},{n}",
                           "-f", "geoparquet", "--type=building", "-o", pq], check=True)
        b = gpd.read_parquet(pq)
        b = b.set_crs(4326) if b.crs is None else b.to_crs(4326)
        b["src"] = b["sources"].astype(str)
        counts = {}
        for label, key in [("google","Google"), ("osm","OpenStreetMap")]:
            sub = gpd.clip(b[b["src"].str.contains(key, case=False, na=False)], poly).reset_index(drop=True)
            counts[label] = len(sub)
            if len(sub):
                sub[["geometry"]].to_file(f"{OUT}/{tag}_{label}.shp")
        done.append(name); print(f"   {name}: {counts}")
    except Exception as e:
        failed.append((name, str(e))); print("   FAILED:", name, "->", e)

print(f"\nDONE {len(done)} | SKIPPED {len(skipped)} | FAILED {len(failed)}")
for n, e in failed: print("  failed:", n, e)
```

```python
import geopandas as gpd, pandas as pd, subprocess, sys, shutil, os
import numpy as np

BND_FILE = "Assam_only.shp"; NAME_COL = "shapeName"
OUT = "footprints"; PQ = "overture_parquet"
os.makedirs(OUT, exist_ok=True); os.makedirs(PQ, exist_ok=True)
OVERTURE = [shutil.which("overturemaps")] if shutil.which("overturemaps") \
           else [sys.executable, "-m", "overturemaps"]

adm = gpd.read_file(BND_FILE)
adm = adm.set_crs(4326) if adm.crs is None else adm.to_crs(4326)
def safe(s): return "".join(c if c.isalnum() else "_" for c in str(s)).strip("_")

RETRY = ["Tinsukia","Chirang","Kamrup","Barpeta","Sonitpur","Udalguri","Golaghat"]
GRID = 3   # 3x3 = 9 tiles per district; raise to 4 if any still fail

for name in RETRY:
    tag = safe(name)
    if os.path.exists(f"{OUT}/{tag}_osm.shp"):
        print("already done:", name); continue
    poly = adm.loc[adm[NAME_COL] == name].geometry.iloc[0].buffer(0)
    w, s, e, n = poly.bounds
    xs = np.linspace(w, e, GRID + 1); ys = np.linspace(s, n, GRID + 1)
    google_parts, osm_parts = [], []
    print(f"{name}: {GRID*GRID} tiles")
    for ix in range(GRID):
        for iy in range(GRID):
            tw, te, ts, tn = xs[ix], xs[ix+1], ys[iy], ys[iy+1]
            tpq = f"{PQ}/{tag}_{ix}{iy}.parquet"
            try:
                if not os.path.exists(tpq):
                    subprocess.run(OVERTURE + ["download", f"--bbox={tw},{ts},{te},{tn}",
                                   "-f","geoparquet","--type=building","-o",tpq], check=True)
                b = gpd.read_parquet(tpq)
                b = b.set_crs(4326) if b.crs is None else b.to_crs(4326)
                b["src"] = b["sources"].astype(str)
                tile_poly = poly.intersection(__import__("shapely").geometry.box(tw,ts,te,tn))
                g = gpd.clip(b[b["src"].str.contains("Google", case=False, na=False)], tile_poly)
                o = gpd.clip(b[b["src"].str.contains("OpenStreetMap", case=False, na=False)], tile_poly)
                if len(g): google_parts.append(g[["geometry"]])
                if len(o): osm_parts.append(o[["geometry"]])
                del b
                print(f"   tile {ix}{iy}: google {len(g)} osm {len(o)}")
            except Exception as e:
                print(f"   tile {ix}{iy} FAILED: {e}")
    if google_parts:
        gpd.GeoDataFrame(pd.concat(google_parts, ignore_index=True), crs=4326).to_file(f"{OUT}/{tag}_google.shp")
    if osm_parts:
        gpd.GeoDataFrame(pd.concat(osm_parts, ignore_index=True), crs=4326).to_file(f"{OUT}/{tag}_osm.shp")
    print(f"   {name} DONE -> google {sum(len(p) for p in google_parts)}, osm {sum(len(p) for p in osm_parts)}")
```

```python
import geopandas as gpd, pandas as pd, subprocess, sys, shutil, os
import numpy as np
from shapely.geometry import box

BND_FILE = "Assam_only.shp"; NAME_COL = "shapeName"
OUT = "footprints"; PQ = "overture_parquet"
OVERTURE = [shutil.which("overturemaps")] if shutil.which("overturemaps") \
           else [sys.executable, "-m", "overturemaps"]

adm = gpd.read_file(BND_FILE)
adm = adm.set_crs(4326) if adm.crs is None else adm.to_crs(4326)
def safe(s): return "".join(c if c.isalnum() else "_" for c in str(s)).strip("_")
def polys_only(g):                      # drop points/lines that break shapefile write
    return g[g.geometry.type.isin(["Polygon", "MultiPolygon"])]

RETRY = ["Tinsukia","Chirang","Kamrup","Barpeta","Sonitpur","Udalguri","Golaghat"]
GRID = 4

for name in RETRY:
    tag = safe(name)
    if os.path.exists(f"{OUT}/{tag}_google.shp") and os.path.exists(f"{OUT}/{tag}_osm.shp"):
        print("already done:", name); continue
    poly = adm.loc[adm[NAME_COL] == name].geometry.iloc[0].buffer(0)
    w, s, e, n = poly.bounds
    xs = np.linspace(w, e, GRID + 1); ys = np.linspace(s, n, GRID + 1)
    gparts, oparts = [], []
    print(f"{name}: {GRID*GRID} tiles")
    for ix in range(GRID):
        for iy in range(GRID):
            tw, te, ts, tn = xs[ix], xs[ix+1], ys[iy], ys[iy+1]
            tpq = f"{PQ}/{tag}_g{GRID}_{ix}{iy}.parquet"
            try:
                if not os.path.exists(tpq):
                    subprocess.run(OVERTURE + ["download", f"--bbox={tw},{ts},{te},{tn}",
                                   "-f","geoparquet","--type=building","-o",tpq], check=True)
                b = gpd.read_parquet(tpq)
                b = b.set_crs(4326) if b.crs is None else b.to_crs(4326)
                b["src"] = b["sources"].astype(str)
                tpoly = poly.intersection(box(tw, ts, te, tn))
                g = polys_only(gpd.clip(b[b["src"].str.contains("Google", case=False, na=False)], tpoly))
                o = polys_only(gpd.clip(b[b["src"].str.contains("OpenStreetMap", case=False, na=False)], tpoly))
                if len(g): gparts.append(g[["geometry"]])
                if len(o): oparts.append(o[["geometry"]])
                del b
            except Exception as ex:
                print(f"   tile {ix}{iy} FAILED: {str(ex)[:60]}")
    if gparts:
        gpd.GeoDataFrame(pd.concat(gparts, ignore_index=True), crs=4326).to_file(f"{OUT}/{tag}_google.shp")
    if oparts:
        gpd.GeoDataFrame(pd.concat(oparts, ignore_index=True), crs=4326).to_file(f"{OUT}/{tag}_osm.shp")
    print(f"   {name} DONE -> google {sum(len(p) for p in gparts)}, osm {sum(len(p) for p in oparts)}")
```

```python
import geopandas as gpd, pandas as pd, subprocess, sys, shutil, os
import numpy as np
from shapely.geometry import box

BND_FILE = "Assam_only.shp"; NAME_COL = "shapeName"
OUT = "footprints"; PQ = "overture_parquet"
OVERTURE = [shutil.which("overturemaps")] if shutil.which("overturemaps") \
           else [sys.executable, "-m", "overturemaps"]
def polys_only(g): return g[g.geometry.type.isin(["Polygon","MultiPolygon"])]

adm = gpd.read_file(BND_FILE)
adm = adm.set_crs(4326) if adm.crs is None else adm.to_crs(4326)

for name in ["Tinsukia", "Kamrup"]:
    poly = adm.loc[adm[NAME_COL] == name].geometry.iloc[0].buffer(0)
    w, s, e, n = poly.bounds
    GRID = 6
    xs = np.linspace(w, e, GRID + 1); ys = np.linspace(s, n, GRID + 1)
    gparts, oparts = [], []
    print(f"{name}: {GRID*GRID} tiles")
    for ix in range(GRID):
        for iy in range(GRID):
            tw, te, ts, tn = xs[ix], xs[ix+1], ys[iy], ys[iy+1]
            tpq = f"{PQ}/{name}_g{GRID}_{ix}{iy}.parquet"
            try:
                if not os.path.exists(tpq):
                    subprocess.run(OVERTURE + ["download", f"--bbox={tw},{ts},{te},{tn}",
                                   "-f","geoparquet","--type=building","-o",tpq], check=True)
                b = gpd.read_parquet(tpq)
                b = b.set_crs(4326) if b.crs is None else b.to_crs(4326)
                b["src"] = b["sources"].astype(str)
                tpoly = poly.intersection(box(tw, ts, te, tn))
                g = polys_only(gpd.clip(b[b["src"].str.contains("Google", case=False, na=False)], tpoly))
                o = polys_only(gpd.clip(b[b["src"].str.contains("OpenStreetMap", case=False, na=False)], tpoly))
                if len(g): gparts.append(g[["geometry"]])
                if len(o): oparts.append(o[["geometry"]])
                del b
            except Exception as ex:
                print(f"   tile {ix}{iy} FAILED: {str(ex)[:50]}")
    if gparts:
        gpd.GeoDataFrame(pd.concat(gparts, ignore_index=True), crs=4326).to_file(f"{OUT}/{name}_google.shp")
    if oparts:
        gpd.GeoDataFrame(pd.concat(oparts, ignore_index=True), crs=4326).to_file(f"{OUT}/{name}_osm.shp")
    print(f"   {name} DONE -> google {sum(len(p) for p in gparts)}, osm {sum(len(p) for p in oparts)}")
```
